```python
nuevos_nombres = {
    'FEC_NOT': 'fechaNotificacion',
    'SEMANA': 'semana',
    'ANO': 'anio',
    'EDAD': 'edad',
    'SEXO': 'sexo',
    'nom_est_f_caso': 'estadoCaso',
    'Nom_upgd': 'nombreUnidad',
    'Departamento_ocurrencia': 'departamento',
    'Municipio_ocurrencia': 'municipio'
}
df_ajuste = df.rename(columns=nuevos_nombres)

df_ajuste['consecutivo'] = 1

df_ajuste.head()
```

## Se crea una nueva columna (grupoEtario)
A partir de las reglas de ciclo de vida dado por el ministerio de salud y protección, se establece lo siguiente:

*   Primera Infancia (0-5 años)
*   Infancia (6 - 11 años)
*   Adolescencia (12 - 18 años)
*   Juventud (14 - 26 años)
*   Adultez (27- 59 años)
*   Persona Mayor (60 años o mas) envejecimiento y vejez.

```python

def asignar_grupo_etario(edad):
    if edad >= 0 and edad <= 5:
        return 'primeraInfancia'
    elif edad >= 6 and edad <= 11:
        return 'infancia'
    elif edad >= 12 and edad <= 18:
        return 'adolescencia'
    elif edad >= 19 and edad <= 26:
        return 'juventud'
    elif edad >= 27 and edad <= 59:
        return 'adultez'
    elif edad >= 60:
        return 'personaMayor'
    else:
        return 'Desconocido'

df_ajuste['grupoEtario'] = df_ajuste['edad'].apply(asignar_grupo_etario)


df_ajuste.head()
```

## Se crea una nueva columna compuesta por año y semana

```python
df_ajuste['anio_semana'] = df_ajuste['anio'].astype(str) + df_ajuste['semana'].astype(str)

df_ajuste['anio_semana'] = pd.to_numeric(df_ajuste['anio_semana'], errors='coerce')

df_ajuste.head()
```
