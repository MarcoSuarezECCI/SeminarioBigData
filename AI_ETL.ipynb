## Imports

```python
import pandas as pd
from google.colab import drive
#import numpy as np
#import matplotlib. pyplot as plt
```

## Proceso de extracción

```python
drive.mount('/content/drive')

file = '/content/drive/MyDrive/data/Articulo/Datos_Dengue.xlsx'

df = pd.read_excel(file)

df.head()
```

## Proceso de Transformación

# Se inicia filtrando los casos para dejar solo los confirmados

```python
df_filtrado = df[~df['nom_est_f_caso'].str.contains('Probable', case=False)]

df_filtrado.head()
```

# Luego se filtra por el departamento objetivo (Atlantico)

```python
df_filtrado_atlantico = df_filtrado[df_filtrado['Departamento_ocurrencia'] == 'ATLANTICO']

df_filtrado_atlantico.head()
```

# Se eliminan las columnas que no son necesarias

```python
columnas_eliminar = [
    'COD_EVE', 'COD_PRE', 'COD_SUB', 'UNI_MED', 'nacionalidad', 'nombre_nacionalidad',
    'COD_PAIS_O', 'COD_DPTO_O', 'COD_MUN_O', 'AREA', 'OCUPACION', 'TIP_SS', 'COD_ASE',
    'PER_ETN', 'GRU_POB', 'nom_grupo', 'GP_DISCAPA', 'GP_DESPLAZ', 'GP_MIGRANT',
    'GP_CARCELA', 'GP_GESTAN', 'sem_ges', 'GP_INDIGEN', 'GP_POBICFB', 'GP_MAD_COM',
    'GP_DESMOVI', 'GP_PSIQUIA', 'GP_VIC_VIO', 'GP_OTROS', 'fuente', 'COD_PAIS_R',
    'COD_DPTO_R', 'COD_MUN_R', 'COD_DPTO_N', 'COD_MUN_N', 'FEC_CON', 'INI_SIN',
    'TIP_CAS', 'PAC_HOS', 'FEC_HOS', 'CON_FIN', 'FEC_DEF', 'AJUSTE', 'CER_DEF',
    'CBMTE', 'FEC_ARC_XL', 'FEC_AJU', 'FM_FUERZA', 'FM_UNIDAD', 'FM_GRADO',
    'confirmados', 'consecutive_origen', 'va_sispro', 'Estado_final_de_caso',
    'Pais_residencia', 'Departamento_residencia', 'Municipio_residencia',
    'Departamento_Notificacion', 'Municipio_notificacion'
]

df_limpiado = df_filtrado_atlantico.drop(columns=columnas_eliminar)

df_limpiado.head()
```
